```python
with plt.style.context("seaborn"):
    fix, axes = plt.subplots(ncols=3,sharey=True, gridspec_kw={
        "width_ratios":[2,1,1],
        "wspace":0
    })
    y='age'
    xs = ['chol', 'trestbps', 'thalach']

    for x, ax in zip(xs, axes):
        ax.scatter(df[x], df[y])
        ax.set_xlabel(x)
        axes[0].set_ylabel(y)
```
